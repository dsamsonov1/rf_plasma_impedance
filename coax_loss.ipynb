```python
import numpy as np
import matplotlib.pyplot as plt


rhoa = rhob = 8e-6 # Ohm/m
eps = 8.3
D = 1
d = np.arange(1e-3, 0.999, 1e-3)
f = 80e6 # Hz
tand = 1e-3

dD = D/d

alpha_c = 1.98e-5*np.sqrt(eps*f)/np.log10(D/d)*(np.sqrt(rhoa)/d+np.sqrt(rhob/D))
alpha_d = 9.08e-8*f*np.sqrt(eps)*tand

alpha = alpha_c+alpha_d

plt.loglog(dD, alpha)
print(alpha_d)
```
